**Line Chart in Matplotlib**

1. Simple Line Plot
    * Illustrates a basic line chart using plt.plot() with x and y arrays (e.g., y = 2×x).
    * Demonstrates how to render with plt.show(). 

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array([1, 2, 3, 4])  # X-axis 
y = x*2  # Y-axis

plt.plot(x, y)  
plt.show()
```

2. Adding Labels & Title
    * Emphasizes the importance of labeling axes (plt.xlabel(), plt.ylabel()) and adding a plot title (plt.title()) for clarity. 

```python
x = np.array([1, 2, 3, 4])
y = x*2

plt.plot(x, y)
plt.xlabel("X-axis")        # Label for the X-axis
plt.ylabel("Y-axis")        # Label for the Y-axis
plt.title("Any suitable title")  # Chart title
plt.show()
```

3. Annotated Line Chart
    * Shows how to use markers (e.g., marker='o') and annotations (plt.annotate()) to label each data point with its coordinates. Includes optional grid lines.

```python
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

plt.figure(figsize=(8, 6))
plt.plot(x, y, marker='o', linestyle='-')

# Add annotations
for i, (xi, yi) in enumerate(zip(x, y)):
    plt.annotate(f'({xi}, {yi})', (xi, yi), textcoords="offset points", xytext=(0, 10), ha='center')

plt.title('Line Chart with Annotations')
plt.xlabel('X-axis Label')
plt.ylabel('Y-axis Label')

plt.grid(True)
plt.show()
```

4. Multiple Line Charts
    * Explains two approaches:
    * Separate figures: create multiple plots in sequential windows using plt.figure().
    * Same axis: overlay multiple lines in one plot (e.g., using a different line style like '-.') to compare datasets.

```python
x = np.array([1, 2, 3, 4])
y = x*2

plt.plot(x, y)
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Any suitable title")
plt.show()  # show first chart

plt.figure()
x1 = [2, 4, 6, 8]
y1 = [3, 5, 7, 9]
plt.plot(x1, y1, '-.')
plt.show()
```

5. Filling Between Lines
    * Demonstrates visualizing areas between two lines using plt.fill_between(), with adjustable transparency via the alpha parameter to highlight differences.

```python
x = np.array([1, 2, 3, 4])
y = x*2

plt.plot(x, y)

x1 = [2, 4, 6, 8]
y1 = [3, 5, 7, 9]

plt.plot(x, y1, '-.')
plt.xlabel("X-axis data")
plt.ylabel("Y-axis data")
plt.title('multiple plots')

plt.fill_between(x, y, y1, color='green', alpha=0.5)
plt.show()
```

6. Multiple Plots on the Same Axis

```python
x = np.array([1, 2, 3, 4])
y = x*2

# first plot with X and Y data
plt.plot(x, y)

x1 = [2, 4, 6, 8]
y1 = [3, 5, 7, 9]

# second plot with x1 and y1 data
plt.plot(x1, y1, '-.')

plt.xlabel("X-axis data")
plt.ylabel("Y-axis data")
plt.title('multiple plots')
plt.show()
```
